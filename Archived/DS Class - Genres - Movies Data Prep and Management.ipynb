```python
import pandas as pd
import numpy as np
import matplotlib

url = "https://raw.githubusercontent.com/wolframbler/dsClass/a118d5b59e36fab22162c3aa8181940c2ca06101/movie_metadata%20-%20clean.csv"

movies_df = pd.read_csv(url, delimiter=",").dropna()

# first, just get a peek at the data:
movies_df.head()
print(movies_df)
```

```python
# print list of column headers
movies_df.columns.tolist()
```

```python
# Parse out individual genres from grouped genres in each row
genre_array = [genre_list.split("|") for genre_list in movies_df['genres']]
distinct_genres = {genre.strip() for genres in genre_array for genre in genres}
distinct_genres.discard('')

# Display individual genres
distinct_genres
```

```python
# Assign 0s to each individual genre
for genre in sorted(distinct_genres):
    movies_df[genre] = np.zeros(len(movies_df), dtype=int)
    
pd.set_option("display.max_columns", 150)
movies_df.head()
```

```python
# Assign the applicable genres with 1s
for idx, row in movies_df.iterrows():
    for genre in row['genres'].split("|"):
        if genre != '':
            movies_df.loc[idx, genre] = 1
```
